```python
def resample_linear(dato_i):
    # resample and fill in with mean value
    dato_i_resampled = dato_i.resample('5min').mean()
    
    for col in dato_i_resampled.columns:
        # valid = (dato_i_resampled.shift(1)[col].notna()) & (dato_i_resampled.shift(-1)[col].notna())
        # dato_i_resampled.loc[valid, col] = dato_i_resampled[col].interpolate(method='linear')
        valid = (dato_i_resampled.shift(1)[col].notna()) & (dato_i_resampled.shift(3)[col].notna())
        dato_i_resampled.loc[valid, col] = 1.5*dato_i_resampled.shift(1)[col]-0.5*dato_i_resampled.shift(3)[col]

    return dato_i_resampled
```

```python
for idx_row in range(0,dist.shape[0]):#dist.shape[0]
    curr_time=dbz_time[idx_row]
    
    curr_ds_0_10=ds_0_10[ds_0_10['time'] == curr_time]
    if curr_ds_0_10.shape[0]==0:
        continue

    curr_dist=dist[idx_row,:]
    station_id=station_sensors[np.where(curr_dist<20000)]

    met_list=list()
    # for sen_var in ["pres", "temp", "relh", "wind", "winv", "rain"]:
    for sen_var in ["rain"]:
        curr_sen_var=[sid for sid in station_id if sid[:4] == sen_var]
        sens_dict[curr_time].append(curr_sen_var)
        curr_dato=curr_ds_0_10[curr_sen_var].values[0]
        curr_dato_nonan = curr_dato[~np.isnan(curr_dato)]  # remove NaN
        if sen_var=="wind":
            if curr_dato_nonan.size == 0:
                met_list.append([np.nan, np.nan, np.nan])
                met_list.append([np.nan, np.nan, np.nan])
            else:
                curr_dato_nonan = np.where(curr_dato_nonan == 888, np.nan, curr_dato_nonan)
                
                curr_dato_nonan_x = np.where(curr_dato_nonan == 777, 0, curr_dato_nonan)
                wind_x=np.sin(np.deg2rad(curr_dato_nonan_x))

                curr_dato_nonan_y = np.where(curr_dato_nonan == 777, 90, curr_dato_nonan)
                wind_y=np.cos(np.deg2rad(curr_dato_nonan_y))
                
                met_list.append([np.nanmin(wind_x), np.nanmean(wind_x), np.nanmax(wind_x)])
                met_list.append([np.nanmin(wind_y), np.nanmean(wind_y), np.nanmax(wind_y)])
        elif sen_var=="rain":
            if curr_dato_nonan.size == 0:
                met_list.append([np.nan, np.nan, np.nan])
            else:
                curr_dato_nonan_nonzero = curr_dato_nonan[curr_dato_nonan > 0]
                if curr_dato_nonan_nonzero.size == 0:                    
                    met_list.append([np.nanmin(curr_dato_nonan), np.nanmean(curr_dato_nonan), np.nanmax(curr_dato_nonan)])
                else:
                    met_list.append([np.nanmin(curr_dato_nonan_nonzero), np.nanmean(curr_dato_nonan_nonzero), np.nanmax(curr_dato_nonan_nonzero)])
            
        else:
            if curr_dato_nonan.size == 0:
                met_list.append([np.nan, np.nan, np.nan])
            else:
                met_list.append([np.nanmin(curr_dato_nonan), np.nanmean(curr_dato_nonan), np.nanmax(curr_dato_nonan)])

    df_met.iloc[idx_row, :]=np.array(met_list).reshape(1,-1)[0]
```

```python
# min max
scaler_X = MinMaxScaler()
scaler_y = MinMaxScaler()

X_train = scaler_X.fit_transform(X_train)
X_val = scaler_X.transform(X_val)
X_test = scaler_X.transform(X_test)

# y_train = scaler_y.fit_transform(y_train)
# y_test = scaler_y.transform(y_test)
y_train = scaler_y.fit_transform(y_train.values.reshape(-1, 1)).flatten()
y_val = scaler_y.transform(y_val.values.reshape(-1, 1)).flatten()
y_test = scaler_y.transform(y_test.values.reshape(-1, 1)).flatten()
```

```python
X_train=X_train.reshape((X_train.shape[0], 2, int((X_train.shape[1]/2))))
X_val=X_val.reshape((X_val.shape[0], 2, int((X_val.shape[1]/2))))
X_test=X_test.reshape((X_test.shape[0], 2, int((X_test.shape[1]/2))))
```

```python
################################################## input layer
input_layer = layers.Input(shape=(X_train.shape[1],X_train.shape[2]))

################################################## hidden layers
# lstm hidden 1
x1 = layers.LSTM(32, return_sequences=False, kernel_regularizer=l1_l2(l1=0.001, l2=0.001))(input_layer) 
x1 = layers.BatchNormalization()(x1) 
x1 = layers.Activation('tanh')(x1)
x1 = layers.Dropout(0.3)(x1)
x1 = layers.Flatten()(x1)

# x12 = layers.LSTM(32,return_sequences=False,kernel_regularizer=l1_l2(l1=0.001, l2=0.001))(x1)
# x12 = layers.BatchNormalization()(x12) 
# x12 = layers.Activation('tanh')(x12)
# x12 = layers.Dropout(0.3)(x12)
# x12 = layers.Flatten()(x12)

# Dense hidden 2
x2 = layers.Flatten()(input_layer)
x2 = layers.Dense(32)(x2) 
x2 = layers.BatchNormalization()(x2)
x1 = layers.Dropout(0.3)(x1)
x2 = layers.Activation('relu')(x2)

# cnn hidden 3
x3 = layers.Conv1D(filters=32, kernel_size=3, padding='same', activation='relu')(input_layer)
x3 = layers.MaxPooling1D(pool_size=2)(x3)
x3 = layers.BatchNormalization()(x3)
x3 = layers.Dropout(0.3)(x3)
x3 = layers.Flatten()(x3)

# concatenate two layers
merged = layers.concatenate([x1, x2, x3], axis=-1)

# hidden layer
x = layers.Dense(16)(merged)
x = layers.BatchNormalization()(x)
x = layers.Activation('relu')(x)

################################################## output layer
# output
output_layer = layers.Dense(1)(x)

# build model
model = models.Model(inputs=input_layer, outputs=output_layer)

model_path=str(computer_user_home)+'/Desktop/5th_exp/'+columns_sid_fixed+'_true'+str(time_forward)+'cumrain_after_'+str(time_forward)+'_minutes_HybridNet_15km.keras'

opt = Adam(learning_rate=0.01)
model.compile(optimizer=opt, loss='mean_squared_error', metrics=['mae'])
callbacks = [
    EarlyStopping(monitor='val_loss', patience=20, verbose=0, restore_best_weights=True),
    ModelCheckpoint(model_path, monitor='val_loss', save_best_only=True,save_weights_only=False,verbose=0),
    ReduceLROnPlateau(monitor='val_loss', factor=0.1, patience=10, min_lr=1e-9, verbose=1)
]

history = model.fit(X_train, y_train, epochs=800, batch_size=64, shuffle=True, validation_data=(X_val, y_val), callbacks=callbacks, verbose=1)
```

```python
def build_model(hp):
    # lstm_units = hp.Int('lstm_units', min_value=16, max_value=64, step=16)
    # dense_units = hp.Int('dense_units', min_value=16, max_value=64, step=16)
    # cnn_filters = hp.Int('cnn_filters', min_value=16, max_value=64, step=16)
    # cnn_kernel_size = hp.Choice('cnn_kernel_size', values=[3, 5])
    # dropout_rate = hp.Float('dropout_rate', 0.1, 0.5, step=0.1)
    final_dense_units = hp.Int('final_dense_units', min_value=8, max_value=32, step=8)
    learning_rate = hp.Float('learning_rate', 1e-4, 1e-2, sampling='log')
    # l1_val = hp.Float('l1', 0.0, 0.01, step=0.001)
    # l2_val = hp.Float('l2', 0.0, 0.01, step=0.001)

    # hp.Int('batch_size', min_value=16, max_value=128, step=16)
    batch_size = hp.Choice('batch_size', [16, 32, 64, 128, 256, 512])
    
    input_layer = layers.Input(shape=(X_train.shape[1], X_train.shape[2]))

    
    # LSTM block
    # x1 = layers.LSTM(lstm_units, return_sequences=False,
    #                  kernel_regularizer=regularizers.l1_l2(l1=l1_val, l2=l2_val))(input_layer)
    # x1 = layers.BatchNormalization()(x1)
    # x1 = layers.Activation('tanh')(x1)
    # x1 = layers.Dropout(dropout_rate)(x1)
    # x1 = layers.Flatten()(x1)
    num_lstm_layers = hp.Int(name='num_lstm_layers', min_value=1, max_value=3, step=1, sampling=None)
    x1 = input_layer
    for i in range(num_lstm_layers):
        return_seq = i < num_lstm_layers - 1
        units = hp.Int(f'lstm_units_{i}', 16, 64, step=16)
        x1 = layers.LSTM(units, return_sequences=return_seq)(x1)
        x1 = layers.BatchNormalization()(x1)
        x1 = layers.Activation('tanh')(x1)
        x1 = layers.Dropout(hp.Float(f'lstm_dropout_{i}', 0.1, 0.5, step=0.1))(x1)
    x1 = layers.Flatten()(x1)

    # Dense block
    x2 = layers.Flatten()(input_layer)
    # x2 = layers.Dense(dense_units)(x2)
    # x2 = layers.BatchNormalization()(x2)
    # x2 = layers.Dropout(dropout_rate)(x2)
    # x2 = layers.Activation('relu')(x2)

    num_dense_layers = hp.Int('num_dense_layers', 1, 4)
    for i in range(num_dense_layers):
        units = hp.Int(f'dense_units_{i}', 16, 128, step=16)
        x2 = layers.Dense(units)(x2)
        x2 = layers.BatchNormalization()(x2)
        x2 = layers.Activation('relu')(x2)
        x2 = layers.Dropout(hp.Float(f'dropout_{i}', 0.1, 0.5, step=0.1))(x2)
    x2 = layers.Flatten()(x2)

    # CNN block
    # x3 = layers.Conv1D(filters=cnn_filters, kernel_size=cnn_kernel_size, padding='same', activation='relu')(input_layer)
    # x3 = layers.MaxPooling1D(pool_size=2)(x3)
    # x3 = layers.BatchNormalization()(x3)
    # x3 = layers.Dropout(dropout_rate)(x3)
    # x3 = layers.Flatten()(x3)
    num_cnn_layers = hp.Int('num_cnn_layers', 1, 3) 
    x3 = input_layer
    for i in range(num_cnn_layers):
        filters = hp.Int(f'cnn_filters_{i}', 16, 64, step=16)
        kernel_size = hp.Choice(f'cnn_kernel_size_{i}', values=[3, 5])
        dropout = hp.Float(f'cnn_dropout_{i}', 0.1, 0.5, step=0.1)
        x3 = layers.Conv1D(filters=filters, kernel_size=kernel_size, padding='same', activation='relu')(x3)
        # x3 = layers.MaxPooling1D(pool_size=1)(x3)
        x3 = layers.BatchNormalization()(x3)
        x3 = layers.Dropout(dropout)(x3)
    x3 = layers.Flatten()(x3)


   
    merged = layers.concatenate([x1, x2, x3], axis=-1)

    # Final hidden + output
    x = layers.Dense(final_dense_units)(merged)
    x = layers.BatchNormalization()(x)
    x = layers.Activation('relu')(x)

    output_layer = layers.Dense(1)(x)

    model = models.Model(inputs=input_layer, outputs=output_layer)

    opt = Adam(learning_rate=learning_rate)
    model.compile(optimizer=opt, loss='mean_squared_error', metrics=['mae'])

    return model


tuner = kt.BayesianOptimization(
    build_model,
    objective='val_loss',
    max_trials=30,
    executions_per_trial=1,
    directory=str(computer_user_home) + '/Desktop/5th_exp/mean_rain_30km_120min_best/' + 'hybridnet_tuning',
    project_name='hybridnet_bayes'
)

log_dir = str(computer_user_home) + '/Desktop/5th_exp/mean_rain_30km_120min_best/' +"logs/fit/" + datetime.datetime.now().strftime("%Y%m%d-%H%M%S")
tensorboard_callback = tf.keras.callbacks.TensorBoard(log_dir=log_dir, histogram_freq=1)

callbacks = [
    EarlyStopping(monitor='val_loss', patience=20, restore_best_weights=True),
    ReduceLROnPlateau(monitor='val_loss', factor=0.1, patience=10, min_lr=1e-9, verbose=0),
    tensorboard_callback
]

tuner.search(
    X_train, y_train,
    epochs=200,
    # batch_size=64,
    # batch_size=tuner.oracle.hyperparameters.Int('batch_size', 16, 128, 16),
    batch_size=tuner.oracle.hyperparameters.get('batch_size'),
    validation_data=(X_val, y_val),
    callbacks=callbacks,
    verbose=1
)

best_model = tuner.get_best_models(num_models=1)[0]
best_hp = tuner.get_best_hyperparameters(1)[0]

model_path = str(computer_user_home) + '/Desktop/5th_exp/mean_rain_30km_120min_best/' + columns_sid_fixed + '_true' + str(time_forward) + 'cumrain_after_' + str(time_forward) + '_minutes_HybridNet_best.keras'
best_model.save(model_path)
```
